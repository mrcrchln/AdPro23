### Scenario
We want to study the agricultural output of several countries, and mainly the differences between the countries China, Spain and Mexico. Our initial thought is that these countries have a huge difference in their agricultural production, as they are very different in size and other political factors. 

We will first start by importing and downloading the data. Secondly we will start with an overall analysis of the world from the years 1980 and 2019. Further we will look more into the countries mentioned above and analyse their evolution. At last we will analyse the choropleth from 2000 and 2019, and the output from the predictor. 

```python
from agros.agros_analysis import Agros

module = Agros()

module.download_data()
```

-----

```python
module.gapminder(1980)
```

```python
module.gapminder(2019)
```

From the gapminder plot we can see that as the fertilizer quantity increase, the output quantity also increase for both years. The size of the dot is also based on the ag_land_quantity, which means that the dot is bigger when the ag_land is bigger. With this in mind, it also seems like the countries with the biggest ag_land_quantity also are the countries with biggest fertilizer quantity and output quantity. 

From 1980 to 2019 it seems like there has been a change from which countries has the highest agricultural production, and the line seems to be even more linear.

```python
print(module.countries_list())

china = 'China'
spain = 'Spain'
mex = 'Mexico'
```

-----

```python
module.area_chart(china, normalize=True)
```

```python
module.area_chart(spain, normalize=True)
```

```python
module.area_chart(mex, normalize=True)
```

In the plots above we can see the differences between the yearly percentage outputs for China, Spain and Mexico. The main differences between the countries is that China has much higher fish output than the other countries, while Mexico have clearly more animal output than the others. All the countries has over 60% of their yearly output in the crop category. Mexico and Spain has had a nearly linear change over the years for the crops, while China started to focus more on fish and less on crop in the middle of the 1980s, and went from around 90% crop in 1960 to around 65% in 2019.

In the plot above we can see the difference in the total output between China, Spain and Mexico. ------

```python
module.variable_correlation()
```

Looking at the heatmap, we can see that there are strong positive correlations between most of the variables.\
output_quantity is highly correlated with crop_output_quantity, animal_output_quantity, ag_land_quantity, capital_quantity, machinery_quantity, livestock_quantity, fertilizer_quantity, and animal_feed_quantity, with coefficients ranging from 0.93 to 1.00. This suggests that these variables tend to increase or decrease together, and that they are likely important factors in determining output_quantity.\
On the other hand, fish_output_quantity has lower correlations with the other variables, ranging from 0.55 to 0.84, suggesting that it may be less closely related to the other variables.Overall, the high positive correlations suggest that these variables are interdependent, and that changes in one variable will likely be accompanied by changes in the other variables. The interpretation of the correlations should be made with caution, as correlation does not necessarily imply causation.

```python
module.chloropleth(2000)
```

```python
module.chloropleth(2019)
```

In the analysis above we can see the tfp (total factor productivity) from the years 2000 and 2019 over the whole world. What we can see is that for the year 2000 the tfp value is under 150 for all countries, and the map is dark purple mostly. For the year 2019 the world plot is much lighter and the tfp value is in general much higher. Iceland has for example gone from dark purple to light yellow, indicating a big growth in the total factor productivity, while Australia has not really shown a big change in the value.
